Create a Dictionary of a text file where every unique word is a key and corresponding frequency is a value

```python
freq = {}

with open(r"E:\5th Sem\Python Class\Day11\c.txt","r") as f:
    
    for line in f:
        words = line.split()
        for word in words:
            if word in freq:
                freq[word] += 1
            else:
                freq[word] = 1
                

print(freq)
```

```python
# pre processing : Bring everything into uniform form where we can apply python operations.

with open(r"E:\5th Sem\Python Class\Day11\c.txt","r") as f:
    
    #print(f.read())
    
    lines = f.readlines()

print(lines)
    
```

```python
# Step 1:

for i in lines:
    i = i.lower()    #here lines.lower() cannot be applied as we cannot apply to List.
    
#print(lines[0])  # No change in the case

length = len(lines)

for i in range(length):
    lines[i] = lines[i].lower()
    
print(lines[0])   # here change occurs
```

```python
print(lines)
```

```python
# Step 2 : Remove unwanted characters

for i in range(length):
    lines[i] = lines[i].replace('\n',"")
    
print(lines)
```

```python
unchars = ['\n','.',',']

for i in range(length):
    for j in unchars:
        lines[i] = lines[i].replace(j," ")

print(lines)
```

```python
#Step 3:  Now convert this entire list into a single string and then do the split() operation to get the word count

strs = ""

for i in range(length):
    strs += lines[i] + " "
    
print(strs)
```

```python
words = strs.split()

print(words)

print(len(words))
```

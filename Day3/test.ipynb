# Control Flow :-> To determine the order of execution


# if statement
# if-else
# if elif else
# nested if

```python
# if Statement

number = 13

guess = int(input("Guess Number:"))

if guess == number:
    print("Success")         #indentation
    print("Perfect")
    
print("Done")
```

```python
# if_else : When we have two possible outcomes

number = int(input("Enter Number:"))

if number % 2 == 0:
    print("Even")
else:
    print("Odd!")
```

```python
# if elif else : When we have multiple conditions

marks = int(input("Enter Marks out of 100"))

if marks > 90:
    print("Grade O")
elif marks > 75:
    print("Grade A")
elif marks > 60:
    print("Grade B")
else:
    print("Fail")
```

```python
# Nested if : When "if" is also inside of "if"

age = 18
citizen = True

if age > 17:
    if citizen == True:
        print("Eligible!")
```

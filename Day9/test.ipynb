# SET  ---Data Structure
 
 Unordered Collection of elements separated by commas and enclosed by curly braces.

```python
primes = {2,3,5,7,11,13}
print(primes)
print(type(primes))
```

```python
#Why Set

for i in primes:
    print(i)
```

```python
print(primes[0])   #it is unordered
```

```python
odds = {1,3,5,7,9,11,13,15}

prime_odds = primes.intersection(odds)

print(prime_odds)
```

```python
all = primes.union(odds)
print(all)
```

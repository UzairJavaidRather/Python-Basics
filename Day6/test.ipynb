# List in List

```python
gatecs = ['ADA',['CLR','SS'],'DM',['RS','SH']]
print(gatecs)
```

```python
print(gatecs[1][1])
```

```python
#using for-loop and print individual elements without printing the lists 

for i in gatecs:
    if isinstance(i,list):
        for j in i:
            print(j)
    else:
        print(i)
```

```python
#Nested lists

gate = ['ADA',['CLR','SS',[12]],'DM',['RS','SH',[15]]]
print(gate)
```

```python
# 3D List

for i in gate:
    if isinstance(i,list):
        for j in i:
            if isinstance(j,list):
                for k in j:
                    print(k)
            else:
                print(j)
    else:
        print(i)
```

```python
#using Functions

def printlist(gate):
    for i in gate:
        if isinstance(i, list):
            for j in i:
                if isinstance(j, list):
                    for k in j:
                        print(k)
                else:
                    print(j)
        else:
            print(i)
        
printlist(gate)
```

```python
#using Recursion to minimize the loops/iterations

def printlist_recursive(gate):
    for i in gate:
        if isinstance(i, list):
            printlist_recursive(i)
        else:
            print(i)

printlist_recursive(gate)
```

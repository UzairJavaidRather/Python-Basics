collection of elements separated by commas and enclosed by parenthesis and non-mutable

```python
r1 = ( "Uzair" ,210309, "Gate")
print(type(r1))

print(r1)
print(r1[0])
print(r1[-1])
```

```python
for i in r1:
    print(i)
```

1. Prompt the user to enter his full name.
2. Remove any extra spaces from the beginning and end and count and print total characters.
3. Print the name in upper and lower case with welcome message.
4. count the actual characters in your name without any spaces.
5. Prompt the user to enter marks and max marks using single input function.
6. Calculate percentage.
7. If percentage > 70 ,then Gate else SSB.

```python
aim = input("Enter your Full name:")
```

```python
print(aim.strip())
print(len(aim.strip()))
```

```python
print("Welcome" + " " + aim.upper())
print("Welcome" + " " + aim.lower())
```

```python
print(aim.replace(" ", ""))   # removes all the spaces from the string "aim"
print(len(aim.replace(" ", "")))   # counts the number of characters in the string "aim" after removing all the spaces
```

```python
marks,mm = input("Enter your marks and maximum marks separated by space:").split()

print(marks)
print(mm)
```

```python
marks = int(marks)
mm = int(mm)

percentage = (marks/mm)*100
print("Your percentage is:", percentage)
```

```python
if percentage > 70:
    print("GATE")
else:
    print("SSB")
```

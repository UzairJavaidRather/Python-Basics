Indexing and Slicing:

# Strings supports character by character accessing using index.


```python
aim = "GATE 2028"
dest = "IITB"

print(aim)
print(dest)

print(aim[0])     # G     ----- indexing from left to right
print(aim[-1])    # 8     ----- indexing from right to left
```

```python
# String slicing

print(aim[0:4])   # GATE  ----- slicing from index 0 to index 3 (4-1) | 0 is inclusive and 4 is exclusive
print(aim[5:])    # 2028  ----- slicing from index 5 to the end of the string

print(aim[ -1 : -4] )   #nothing will be printed because we are slicing from right to left, so the starting index should be lower than the ending index
print(aim[ -4 : -1] )     # we always slice from left to right, so the starting index should be less than the ending index
```

# Indexing : print entire string or characters 
# Slicing : print a sequence of string

# Strings are immutable i.e no change occurs

```python
#e.g

aim[0] = "I"   # this will give an error because strings are immutable, we cannot change the value of a string once it is created   
#but if 
aim = "IIT"
print(aim)   # this will work because we are creating a new string and assigning it to the variable aim, we are not changing the value of the original string "GATE 2028"
```

# STRING OPERATIONS

```python
# Concatenation of strings

aim = "GATE"
dest = "IITB"
print(aim + " " + dest)   # GATE IITB  ----- concatenation of strings using the + operator
```

```python
# Repetition of strings

print(aim * 3)   # GATEGATEGATE  ----- repetition of strings using the * operator
```

```python
# Membership operator
print("G" in aim)   # True  ----- because "G" is present in the string "GATE"
print("Z" in aim)   # False  ----- because "Z" is not present in the string "GATE"
```

```python
# Comparison of strings

print("GATE" == "GATE")   # True  ----- because both the strings are same
print("GATE" == "gate")   # False  ----- because both the strings are different (case sensitive)
print("GATE" > "gate")  # False  ----- because "G" is less than "g" in ASCII values
```

# STRING METHODS

```python
# Upper case and lower case

print(aim.upper())   # GATE  ----- converts all the characters in the string to upper case
print(aim.lower())   # gate  ----- converts all the characters in the string to lower case

print(aim.islower())  # False  ----- because there is at least one upper case character in the string
```

```python
# Counting the number of occurrences of a character in a string

print(aim.count("A"))   # 1  ----- because there is one "A" in the string "GATE"
print(aim.count("Z"))   # 0  ----- because there is no "Z" in the string "GATE"
```

# Strip() :

It removes extra spaces in brackets or unspecified characters from the beginning and end of a string but not in the middle. 

```python
aim = " GATE 2028 "
print(len(aim))   # 11  ----- because there are 11 characters in the string " GATE 2028 " including the spaces

print(aim.strip())  # GATE 2028  ----- removes the leading and trailing spaces
print(len(aim.strip()))  # 9  ----- because there are 9 characters in the string "GATE 2028" after removing the leading and trailing spaces
```

```python
dest = "###IIT#B###"
print(dest.strip("#"))  # IITB  ----- removes the leading and trailing # characters but not the ones in the middle

#SCOPE :    Strip() => Both sides
#           lstrip() => Left side
#           rstrip() => Right side            
```

# FIND and REPLACE

```python
aim = "GATE 2028"

print(aim.find("2028"))   # 5  ----- because the substring "2028" starts at index 5 in the string "GATE 2028"
print(aim.find("IITB"))   # -1  ----- because the substring "IITB" is not present in the string "GATE 2028"
```

```python
print(aim.replace("2028", "2029"))   # GATE 2029  ----- replaces the substring "2028" with "2029" in the string "GATE 2028"

print(aim)   # GATE 2028  ----- the original string is not changed because strings are immutable, the replace() method returns a new string with the replacement done, it does not change the original string
```

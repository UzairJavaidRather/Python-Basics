Write a python program that asks the user to enter a sentence and perform following tasks:

1. Remove leading and trailing spaces.
2. Create a new string formed by taking the first character of each word of the sentence.
3. Convert the new string to upper and lower case.
5. Print the generated string.

```python
# Ask the user for a sentence
sentence = input("Enter a sentence: ")

# Remove leading and trailing spaces
trimmed = sentence.strip()
print("Trimmed sentence:", trimmed)

# Create a new string from the first character of each word
words = trimmed.split()
first_chars = "".join(word[0] for word in words if word)

# Convert the generated string to upper and lower case
upper_string = first_chars.upper()
lower_string = first_chars.lower()

# Print the generated string and its converted forms
print("Generated string:", first_chars)
print("Upper case:", upper_string)
print("Lower case:", lower_string)
```

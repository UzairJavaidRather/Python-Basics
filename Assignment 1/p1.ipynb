Problem 1: Travel Cost & Distance Converter:

1. Prompt the user to enter distance to travel.
2. Ask the user to choose unit: 'K' for kilometers or 'M' for miles.
3. If 'M', convert miles to kilometers (1 mile = 1.609 km).
4. Prompt for fuel efficiency (km per liter).
5. Prompt for fuel price per liter.
6. Compute total fuel required and total travel cost.
7. Display trip category based on distance.
8. Display whether journey is expensive or affordable.

```python
dist = float(input("Enter the distance in kilometers: "))
```

```python
print(dist)
```

```python
unit = input("Choose unit ('K' for kilometers, 'M' for miles): ").strip().upper()
print(unit)
```

```python
if unit == 'M':
        distance_km = dist * 1.609
elif unit == 'K':
        distance_km = dist
else:
        print("Invalid unit choice. Defaulting to kilometers.")
        distance_km = dist
```

```python
print(distance_km)
```

```python
fuel_efficiency = float(input("Enter fuel efficiency (km per liter): "))
print(fuel_efficiency)
```

```python
fuel_price = float(input("Enter fuel price per liter: "))
print(fuel_price)
```

```python
total_fuel_required = distance_km / fuel_efficiency
total_travel_cost = total_fuel_required * fuel_price
```

```python
print("Total distance:" + str(distance_km) + " " + "km")
print("Total fuel required:" + str(total_fuel_required) + " " + "liters")
print("Total travel cost:" + str(total_travel_cost) + " " + "rupees")
```

```python
if distance_km < 50:
        category = "Short Trip"
elif 50 <= distance_km <= 200:
        category = "Medium Trip"
else:
        category = "Long Trip"
        
print("Trip Category: " + category)
```

Let's define a threshold for "expensive"  to be above 10000 rupees

```python
cost_threshold = 10000

if total_travel_cost > cost_threshold:
        status = "Expensive"
else:
        status = "Affordable"
        
print("Trip Status: " + status)
```

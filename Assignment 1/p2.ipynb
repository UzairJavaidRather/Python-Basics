Problem 2: Mathematical Computation & Series:

1. Write an iterative function to compute product of first n natural numbers.
2. Write a recursive function for the same.
3. Write a function for geometric progression term (a * r^n).
4. Compute sum of squares from 1 to n.
5. Compute sum of x^i / i from i=1 to n.
6. Compare computed values and comment on results.

```python
n = int(input("Enter a value for n: "))
x = float(input("Enter a value for x: "))
a = float(input("Enter a value for a (GP start): "))
r = float(input("Enter a value for r (GP ratio): "))

print(n)
print(x)
print(a)
print(r)
```

Iterative Product

```python
def iter(n):
    result = 1   #start from 1
    for i in range(1, n + 1):
        result *= i
    return result
```

Recursive Product

```python
def rec(n):
    if n == 0:
        return 1
    else:
        return n * rec(n - 1)
```

Geometric Progression

```python
def gp(a,r,n):
    return a * (r ** n)
```

Sum of Squares

```python
def sum(n):
    total = 0
    for i in range(1, n + 1):
        total = total + i*i
    return total
```

Series ( x ^ i / i)

```python
def series(x,n):
    total = 0
    for i in range(1, n + 1):
        total = total + (x ** i) / i
    return total
```

Printing

```python
print("Iterative Factorial of n:", iter(n))
print("Recursive Factorial of n:", rec(n))
print("n-th term of GP:", gp(a,r,n))
print("Sum of squares of first n natural numbers:", sum(n))
print("Sum of series x^i/i for i=1 to n:", series(x,n))
```

1. Iterative and recursive product give the same result.
2. Iterative is faster and uses less memory.
3. Recursive method is easier to understand but slower for larger n.
4. Sum function increases as n increases.
5. GP term grows fast if r>1.

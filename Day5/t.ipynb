# List Methods 

1. append() add single element at the last to the list
                gatecs.append("CN)

```python
gatecs.append("CN")
print(gatecs)
```

```python
#pop() method is used to remove the last element from the list

gatecs.pop()
print(gatecs)
```

```python
# extend() method is used to add multiple elements to the list at last
gatecs.extend(["CN","DBMS"])
print(gatecs)
```

```python
# remove() method is used to remove the first occurrence of the specified element from the list
# insert() method is used to insert an element at a specific position in the list

gatecs.remove("OS")
print(gatecs)

gatecs.insert(2,"EM and DM")
print(gatecs)
```

```python
#list inside list

gate = [["DAA","OS","TOC"],["CN","DBMS"]]
print(gate)
```

```python
gatecs.insert(1,15)
gatecs.insert(3,10)
gatecs.insert(5,15)
gatecs.insert(7,10)
gatecs.insert(9,15)
gatecs.append(10)
print(gatecs)
```

```python
print(gatecs[2])
```

```python
# using for - loop
#in is used to check if an element is present in the list or not .....operator
#i is used to access the elements of the list

for i in gatecs:
    print(i)
```

```python
# using while loop

i = 0
while i < len(gatecs):
    print(gatecs[i])
    i= i + 1
```

1. Prompt the user "Enter your Weight (in kgs)" and record weight
2. Prompt the user "what is your preferred unit of height? Type "F" for feet and "M" for meters and record preference
3. If user says "F" then  prompt user "you will enter your height given as feet and inches.First enter feet and record feet and then prompt user "Now enter the inches" and record inches.
4. If user says "M" then prompt user " What is your height in meters" and record in meters.
5. If user had choosen "F" then convert height in meters.
6. Compute BMI using the formula.
7. Depending on the value of BMI report the user's type given by the following table:
            

```python
weight=float(input("Enter your weight in kgs:"))
print(weight)
```

```python
prefheight=input("Enter preferred height in feet or meters :F or M")
print(prefheight)
```

```python
if prefheight == "F":
    feet,inch = input("Enter height in feet and inches:").split()
    feet=float(feet)
    inch=float(inch)
    inches= (feet*12) + inch
    centi = inches * 2.54
    height = centi/100
    print(height) 
elif prefheight == "M":
    height=input("Enter height in meters:")
    height=float(height)
    print(height)
```

```python
BMI = weight / (height*height)
print(BMI)
```

```python
if BMI < 18.5:
    print("Underweight")
elif BMI < 25:
    print("Normal")
elif BMI < 30:
    print("Overweight")
elif BMI > 30:
    print("Very over weight")
```

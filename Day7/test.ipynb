```python
def matrix_multiply(A, B):
    result = [[0, 0], [0, 0]]
    
    size = len(A)
    
    for i in range(size):
        for j in range(size):
            for k in range(size):
                result[i][j] += A[i][k] * B[k][j]
    
    return result


def matrix_power(A, n):
    
    if n == 0:
        return (print("Identity matrix"))  
    
    if n == 1:
        return A
    
    
    prev = matrix_power(A, n - 1)
    return matrix_multiply(prev, A)



A = [[1, 1], [1, 0]]

n = int(input("Enter power: "))

print(matrix_power(A, n))
```

```python
def matrixpower(A,n):
    #B = A    This is a shallow copy ,[:]=> means copy both rows and columns
    B = [i [:] for i in A]
    C = [i[:] for i in A]
    for i in range(n-1):
        C[0][0] = A[0][0] * B[0][0] + A[0][1] * B[1][0]
        C[0][1] = A[0][0] * B[0][0] + A[0][1] * B[1][1]
        C[1][0] = A[1][0] * B[0][0] + A[1][1] * B[1][0] 
        C[1][1] = A[1][0] * B[0][1] + A[1][1] * B[1][1]
    
        B = [ i[:] for i in C]
    
    return C
    

A = [[1,1],[1,0]]

print(matrixpower(A,1))
print(matrixpower(A,2))
print(matrixpower(A,3))
```

```python
for n in range(10):
    result = matrixpower(A,n)
    print(n,result)
```

```python
for n in range(10):
    result = matrixpower(A,n)
    print(n,result[0][0])
```

```python
%%timeit -n1 -r10

mylist = list(range(100000))

for i in range(100000):
    x = 1113 in mylist
```

```python
%%timeit -n1 -r10

mytuple = tuple(range(100000))

for i in range(100000):
    x = 1113 in mytuple
```

```python
%%timeit -n1 -r10

myset = set(range(100000))

for i in range(100000):
    x = 1113 in myset
    
```

Question - Why is the set faster than the list and tuple for membership testing?

Answer - The set is faster than the list and tuple for membership testing because sets in Python are implemented as hash tables, which allow for average-case O(1) time complexity for membership tests. In contrast, lists and tuples require O(n) time complexity for membership tests, as they need to iterate through each element to check for the presence of the item.

# Dictionary 

Data structure that stores keys as well as values in the form of (Keys : Values).

```python
s1 = {"Name" : "Uzair","Rank" : 100,"Institute":"IITB"}

print(s1)
print(type(s1))
```

```python
print(s1["Name"])    # Keys allow you to index data items directly using strings which are more interpretable and more relative .
print(s1["Rank"])    # inorder to qualify for a key the data item must be immutable.
```

```python
s1["Batch"] = "CSE"   # adding of key and values
print(s1["Batch"])
```

Methods on dictionary:

```python
print(s1.keys())  # keys() will list all the keys in our dictionary
```

```python
for i in s1:
    print(i,":",s1[i])
    
```

```python
s2 = {"Name":"Samin","Rank":101,"Institute":"IITD","Branch":"CSE"}
print(s2)

GQ = [s1,s2]

print(GQ)
print(type(GQ))
```

```python
print(type(GQ[0]))
print(type(GQ[0]["Name"]))
```

```python
print(GQ)
print(GQ[0])
print(GQ[0],"Name")
```

```python
for i in GQ:
    if isinstance(i, dict):
        for j in i:
            print(j,":",i[j])
    else:
        print(i)
        
```

```python
# Prompt the user to enter your fav IIT : reply like "I like IITB"

IITS = {"IITB":"Get rank 1 to 99","IITD":"Get rank 100 to 200","IITM":"Get rank 200 to 500"}

reply = input("enter a reply:")
print(reply)

words = reply.split()
print(words)    # in the form of list

fav = words[-1]

 
if fav in IITS:
    print(IITS[fav])
else:
    print("IIT not found")
```

```python
reply = input("enter a reply:")

output = ""

word = reply.split()

for i in word:
    if i in IITS:
        output += IITS[i]
    else:
        output = ""
        
print(output) 
        
```

```python
# Can be done in less lines using "get()" ---fetch corresponding values based on the key

reply = input("enter a reply:")

output = ""

word = reply.split()

for i in word:
    output = IITS.get(i,"")   #it checks if the key is present in the dictionary or not, if it is present then it fetches the value otherwise it returns the default value which is "" in this case.
        
print(output)
        
```
